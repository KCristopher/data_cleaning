```python
def get_iterab_of_dictions_given_file_of_jsons(file_name : str) -> List[dict] :

    data_list = []
    current_json_lines = []

    with open(file_name, 'r') as file_js:
        for line in file_js:
            # Strip any leading or trailing whitespace from the line
            line = line.strip()

            if line.startswith('{') and line.endswith('}'):
                # This line contains a complete JSON object
                current_json_lines.append(line)
                data = json.loads(''.join(current_json_lines))
                data_list.append(data)
                current_json_lines = []
            elif line.startswith('{'):
                # This line starts a JSON object that spans multiple lines
                current_json_lines.append(line)
            elif line.endswith('}'):
                # This line ends a JSON object that spans multiple lines
                current_json_lines.append(line)
                data = json.loads(''.join(current_json_lines))
                data_list.append(data)
                current_json_lines = []
            else:
                # This line is part of a JSON object that spans multiple lines
                current_json_lines.append(line)

    # Now you have a list of JSON objects in data_list
    return data_list
```

```python
def concat_diction(dictions_iterab) -> dict :
    
    """
    Takes a list of dictionaries and output
    one big dictionary with the unique keys of
    the original dictionaries as keys and the combined
    values sharing the same key as values.
    The keys are strings and the values are lists.
    
    
    Parameters
    ----------
    dictions_iterab : list. List of dictionaries, 
                      which we want to combine into one.
                      
    Returns
    -------
    dict.
    """
    
    keys = [k for di in dictions_iterab for k in di.keys()]
    values = [v for di in dictions_iterab for v in di.values()]
    keys_uniq = list(set(keys))
    values_empty = [ [ ] for _ in range(len(keys_uniq))]
    big_diction = dict(zip(keys_uniq, values_empty))
    pairs = [*zip(keys, values)]
    
    for p in pairs :
        key = p[0]
        value = p[1]
        big_diction[key].append(value)
    
    return big_diction
```

```python
def make_diction_into_frame(diction) :
    
    return pd.DataFrame.from_dict(diction)
    
```

```python
df['num_char_quartile'] = pd.qcut(df.num_char, 4)
df['num_char_deciles'] = pd.qcut(df.num_char, 10)
df['likes_quartiles'] = pd.qcut(df.likes, 4, duplicates = 'drop')
df['likes_deciles'] = pd.qcut(df.likes, 10, duplicates = 'drop')
df['likes_cat'] = pd.cut(df.likes, 10)
df['num_char_cat'] = pd.cut(df.num_char, 5)
```
